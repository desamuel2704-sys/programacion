```python

# Solicitud de datos al usuario transformados a int


principal_vp = int(input("Introduce el monto inicial invertido (vP): "))

tasa_interes_i = int(input("Introduce la tasa de interés anual (%) (i): "))

años_n = int(input("Introduce el número de años (n): "))

 

# Cálculo del Valor Futuro (Interés Simple)

# Dividimos la tasa entre 100 para convertir el número entero a decimal

valor_futuro = principal_vp * (1 + (tasa_interes_i / 100) * años_n)

 

print("-" * 30)

print(f"Al final de {años_n} años, tu inversión valdrá: ${valor_futuro:.2f}")

```
